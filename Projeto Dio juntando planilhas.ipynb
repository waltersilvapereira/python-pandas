```python
import pandas as pd
```

```python
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
#alterar um tipo de dado. 
df['LojaID'] = df["LojaID"].astype("object")
```

```python
df.isnull().sum()
```

```python
#substituir os valores nulos pelas medias
df["Vendas"].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Vendas"].mean()
```

```python
#apagar linhas com valores nulos
df.dropna(inplace=True)
```

```python
#apagar linhas com valores nulos em uma coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#apagar linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

```python
#criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#maiores receitas
df.nlargest(3,"Receita")
```

```python
#menores receitas
df.nsmallest(3,"Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

```python
df.dtypes
```

```python
#agrupamento da receita por ano 
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_venda"] = df["Data"].dt.year
```

```python
df["Mes_venda"] = df["Data"].dt.month
```

```python
df["Dia_venda"] = df["Data"].dt.day
```

```python
df.head()
```

```python
#metodo sample para extrair uma amostragem 
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data"].min()
```

```python
#Calculando a diferença de dias
df["diferença_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#Criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
#Filtrando as vendas de 2019 do mes de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019)&(df["Data"].dt.month == 3)]
```
